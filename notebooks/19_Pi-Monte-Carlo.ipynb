# Determinando Pi con Monte Carlo

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%config InlineBackend.figure_format='retina'
```

```python
np.random.seed(43)
x = np.random.uniform(0, 1, 3000)
np.random.seed(35)
y = np.random.uniform(0, 1, 3000)
```

¿Qué pasa si uso la misma semilla para ambas listas?

```python
fig, ax = plt.subplots()
ax.scatter(x, y, s=0.7)
ax.set_aspect(1)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
sns.histplot(x, bins=20)
plt.xlabel('x')
plt.show()
```

```python
mask = np.where(np.sqrt(x**2 + y**2) <=1)

x_in = x[mask]
y_in = y[mask]

x_out = np.delete(x, mask) # numpy delete para eliminar los que estan dentro del ciculo
y_out = np.delete(y, mask)
```

```python
# circle
theta = np.linspace(0, np.pi/2, 100)
x_line = np.cos(theta)
y_line = np.sin(theta)

fig, ax = plt.subplots()
ax.scatter(x_in, y_in, s=0.7)
ax.scatter(x_out, y_out, s=0.7)
ax.plot(x_line, y_line, 'k')
ax.set_aspect(1)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

$$\frac{\pi}{4} ≈
\frac{n_{\rm in}}{n_{\rm total}}$$

```python
pi_MC = 4*len(x_in)/len(x)
print(f"Pi MC = {pi_MC:.4f}")
```
