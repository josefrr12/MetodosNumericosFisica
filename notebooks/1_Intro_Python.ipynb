## NumPy
Numerical Python. Calculos numericos vectorizados. ¿Qué pasa si multiplicamos dos lista?

```python
list_a = [1, 2, 3, 10, 12, 13]
list_b = [2, 4, 6, 20, 25, 26]

list_a * list_b #error
```

```python
list_c = []
for i in range(len(list_a)):
    list_c.append(list_a[i] * list_b[i])

list_c
```

```python
list_c = [list_a[i]*list_b[i] for i in range(len(list_a))]
list_c
```

```python
import numpy as np # alias np

array_a = np.array(list_a)
array_b = np.array(list_b)
```

```python
array_c = array_a * array_b
array_c
```

## Matplolib
Libreria para hacer graficas

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0.01, 10, 0.01)
plt.figure(figsize=(5,5))
plt.plot(x, np.floor(x), label='Funcion piso')
plt.plot(x, np.ceil(x), label='Funcion techo')
plt.legend(fontsize=12)
plt.xlabel(r'$x$')
plt.ylabel('Funcion')
plt.title('Mis funciones')
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.grid(True)
plt.show()
```
