# Clases
Vamos a crear la "idea" abstracta de gato. Un gato tiene caracteristica generales y propias. Ademas puede realizar "acciones"

```python
class Gato:
    # atributos
    especie = "Felis catus"
    clase = "Mammalia"

    # constructor
    def __init__(self, nombre, edad):
        # atributo
        self.nombre = nombre
        self.edad = edad

    # metodo
    def maullar(self):
        print( f"{self.nombre} dice miau!")
```

Una "instancia", del ingles instance, es un ejemplo de una clase abstracta. Instantiate = crear una "instancia", un ejemplo.

```python
# Crear "instancias"
gato1 = Gato("Bruno", 15)
gato2 = Gato("Isis", 17)
```

```python
print(gato1.especie)
```

```python
print(gato1.nombre)
print(gato2.nombre)
```

```python
gato1.maullar()
```

```python
gato2.maullar()
```

# Ejemplos de clases en python

```python
import numpy as np
A = np.array([[1, 4, 5],[5, 7, 9],[9, 10, 11], [12, 14, 1]])
print(type(A))
```

## Atributos numpy array

```python
print(A.size)
print(A.shape)
print(A.dtype)
```

## Metodos

```python
print(A.min())
print(A.min(axis=0))
print(A.min(axis=1))
# suma
print(A.sum(axis=0))
print(A.sum(axis=1))
print(A.sum())
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 1, 0.01)
y = x**0.5
```

```python
# funciones
plt.figure()
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Enfoque funcional')
plt.show()
```

```python
# objetos
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Enfoque con objetos")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid()
ax.set_xticks([0.1, 0.7])
ax.set_xticklabels([':)', '(:'])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Enfoque con objetos")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid()
ax.set_xscale('log')
ax.set_yscale('log')
plt.show()
```

```python
x = np.linspace(0, 6*np.pi, 201)
y = np.linspace(0, 6*np.pi, 201)
X, Y = np.meshgrid(x, y)
Z = np.sin(X+Y)
# graficar
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X,Y,Z)
ax.set_title('Figura 3D')
ax.view_init(elev=45) # angulo vision
plt.show()
```
