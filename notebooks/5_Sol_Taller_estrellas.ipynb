```python
df_stars['Superficial Area'] = 4*np.pi*df_stars['Radius(R/Ro)']**2
df_stars.head(5)
```

```python
plt.figure()
# usamos seaborn
sns.scatterplot(data=df_stars, x='Temperature (K)', y='Luminosity(L/Lo)', hue='Star type')
# escala logaritmica en ambos ejes
plt.xscale('log')
plt.yscale('log')
# seleccionar los ejes
ax = plt.gca()
# invertir el eje x
ax.invert_xaxis()
plt.legend(fontsize=10)
plt.show()
```

```python
df_stars['Flux at surface'] = df_stars['Luminosity(L/Lo)']/df_stars['Superficial Area']
df_stars.head(10)
```

```python
plt.figure()
sns.scatterplot(data=df_stars, x='Temperature (K)', y='Flux at surface', hue='Star type')
plt.xscale('log')
plt.yscale('log')
plt.legend(loc='lower right', fontsize=8)
plt.show()
```

```python
df_main_seq = df_stars[df_stars['Star type']=='Main Sequence']
df_main_seq.head(5)
```

```python
# linealizar las variables
log_temp = np.log(df_main_seq['Temperature (K)'])
log_flux = np.log(df_main_seq['Flux at surface'])
```

```python
m, c = np.polyfit(log_temp, log_flux, 1)
alpha = np.exp(c)
n = m
print(m, c)
print(alpha, n)
```

```python
X_plot = np.linspace(np.min(log_temp), np.max(log_temp),10)
y_plot = m*X_plot + c
```

```python
plt.scatter(log_temp, log_flux, label='datos')
plt.plot(X_plot, y_plot, color='r', label='ajuste')
plt.xlabel(r'$\ln(T)$')
plt.ylabel(r'$\ln(f)$')
```

```python
y_fit = m*log_temp + c
res = log_flux - y_fit
sns.histplot(res)
```

```python
Lsun = 3.828e26 # W
Rsun = 6.957e8 # m
conversion = Lsun/Rsun**2
print(alpha*conversion)
```
