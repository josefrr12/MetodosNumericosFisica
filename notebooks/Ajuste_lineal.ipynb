# Taller con estrellas

```python
# importar las librerias necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Ajuste lineal

```python
# Generar datos con un poco de ruido (luego estudiaremos como generar numeros aleatorios)
np.random.seed(42)
alpha_true = 2
n_true = 2
x_data = np.linspace(1, 5, 50) # generar una lista de numeros con espaciamiento lineal
y_noise = 1.5*np.random.normal(size=x_data.size)
y_data = alpha_true * (x_data ** n_true) + y_noise
y_data = np.maximum(y_data, 1e-6) # numeros positivos
```

```python
plt.figure()
plt.scatter(x_data, y_data)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Los datos no siguen una relacion lineal. Parece una ley de potencias $ y = \alpha x^n$. Para confirmar hallamos el logaritmo.

```python
log_x = np.log(x_data)
log_y = np.log(y_data)
```

```python
plt.figure()
plt.scatter(x_data, y_data)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#ajuste lineal (y = mx + c)
m, c = np.polyfit(log_x, log_y, 1)

# c = log(alpha) => alpha = exp(c)
n_fit = m
alpha_fit = np.exp(c)

print(f"Verdaderos: alpha = {alpha_true:.4f}, n = {n_true:.2f}")
print(f"Ajuste: alpha = {alpha_fit:.2f}, n = {n_fit:.2f}")
```

```python
fig, ax = plt.subplots()

# Datos
ax.scatter(x_data, y_data, label='Datos')

# ajuste
x_fit = np.linspace(min(x_data), max(x_data), 100)
y_fit = alpha_fit * (x_fit ** n_fit)
ax.plot(x_fit, y_fit, label=r'Ajuste: $y = {0:.2f}x^{{{1:.2f}}}$'.format(alpha_fit,n_fit), color='C1', linewidth=2)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.set_xscale('log')
ax.set_yscale('log')
plt.show()
```

* Hacer una grfica con las variables originales
* Hallar el error del ajuste.

```python
y_fit = alpha_fit*x_data**n_fit
print(len(y_fit)==len(y_data))
```

```python
error_absoluto = np.abs(y_data - y_fit)
error_relativo = error_absoluto/y_data
error_cuadradratico = (y_data - y_fit)**2
```

```python
#### grafica error absoluto ####
plt.figure()
plt.scatter(y_data, error_absoluto)
plt.xlabel('x datos')
plt.ylabel('error absoluto')
plt.show()
#### grafica error absoluto ####
plt.figure()
plt.plot(x_data, error_relativo)
plt.xlabel('x datos')
plt.ylabel('error absoluto')
plt.show()
#### grafica error cuadratico ####
plt.figure()
plt.plot(x_data, error_cuadradratico)
plt.xlabel('x datos')
plt.ylabel('error absoluto')
plt.show()
```
