Power spectral density  $S_n(f)$ (PSD) y amplitude spectral density (ASD)
$$S_n(f) = ASD^2$$
La razon señal ruido esperada
$$\langle\rho^2\rangle = \frac{4G F_+^2}{2 c^3\pi^2 d^2}\int_2^{f_{isco}} \frac{1}{f^2 S_n}\frac{dE}{df}df $$
donde es $F_+^2 = 1/5$, $d$ es la distancia a la fuente, se tomara como 40 Mpc y $dE/df = -dE_{orb}/df$, $E_{orb}$ es la energia total en una orbita circular,
$$ E_{orb} = -\frac{1}{2}\frac{G m_1 m_2}{R}$$
Ley de Kepler,
$$ \frac{\omega}{2} = \Omega = \sqrt{\frac{G M}{R^3}}$$
Recordar $\Omega$ es la frecuencia angular orbital y $\omega$ es la frecuencia angular de la onda gravitational $\omega=2\pi f$. La masa total de la binaria $M=m_1+m_2$

```python
ligo_optimized_asd = np.genfromtxt('NSNS_Opt.txt')
freqs = ligo_optimized_asd[:,0]
ASD = ligo_optimized_asd[:,1]
```

```python
plt.loglog(freqs, ASD)
plt.xlabel('Frequency (Hz)')
plt.ylabel(r'ASD (1/Hz$^{1/2}$)')
```
