# Metodo de biseccion para hallar ceros de una funcion

Este metodo esta basado en el teorema del valor medio


```python
# libreria numerica
import numpy as np
# libreria graficas
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.exp(x - np.sqrt(x)) - x

def V_eff(r, E = -0.1):
    return 1/r**2-1/r - E

def biseccion(f, x0, x1, kmax=200, tol=1.e-13):
    f0 = f(x0)
    for k in range(1,kmax):
        x2 = (x0 + x1) / 2
        f2 = f(x2)
        if np.abs(f2) < tol:
          x2new = x2
          break
        if f0 * f2 < 0:
            x1 = x2
        else:
            x0, f0 = x2, f2
        x2new = (x0 + x1) / 2
        xdiff = abs(x2new - x2)
        if abs(xdiff / x2new) < tol:
            break
        else:
            x2new = None
    print(f"{k} iteraciones")
    return x2new
```

```python
x = np.arange(0.1, 5, 0.01)
f_val = f(x)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, f_val)
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.grid(True)
plt.show()
```

```python
sol = biseccion(f, 0.5, 1.51)
print(sol)
print(f(sol))
```

```python
r = np.arange(0.1,20, 0.01)
V_val = V_eff(r)
```

```python
plt.plot(r, V_val)
plt.ylim(-0.2, 0.3)
plt.xlabel(r'$r$')
plt.ylabel(r'$V_{eff} - E$')
plt.grid(True)
plt.show()
```

```python
sol_Veff = biseccion(V_eff,1, 5)
print(sol_Veff)
print(V_eff(sol_Veff))
```
